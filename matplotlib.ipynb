```python
batsman=pd.read_csv('metplotlib/sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashed',linewidth=3,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='Green',marker='D',markersize=10,label='Rohit')
plt.title("Rohit Sharma vs Virat Kohli Carrer Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Score')

plt.legend()
plt.grid()
plt.show()
```

```python
df=pd.read_csv('metplotlib/batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(np.array(df.shape[0]),df['2016'])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2017'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2015'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
children=[10,20,30,40,50]
color=['blue,blue,blue,blue,blue','red,red,red,red','black,black,black','green,green','yellow']
plt.bar(color,children,color='pink')
plt.xticks(rotation='vertical')
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')
plt.legend()
plt.show()
```
